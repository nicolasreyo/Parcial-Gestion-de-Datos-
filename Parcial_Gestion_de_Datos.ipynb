# Punto 6


```python
#Se crea un nuevo dataframe netamente con las filas donde su writing score se encuentren por encima del percentil 85
df_perc_85 = df[df['writing score'] > df['writing score'].quantile(0.85)]
```

```python
parents_education = df_perc_85.groupby('parental level of education').size().sort_values(ascending = False)
parents_education = parents_education.reset_index().rename(columns={0: 'Cantidad'})
```

```python
parents_education['Porcentaje'] = (parents_education['Cantidad'] / sum(parents_education['Cantidad']))* 100
parents_education
```

#Punto 7

```python
#Filtro del data frame por estudiantes que hayan sacado 90 o mas en las tres evaluaciones
df_higher_90 = df[(df['math score'] >= 90) & (df['reading score'] >= 90) & (df['writing score'] >= 90)]
```

```python
df_higher_90.head()
```

```python
df_higher_90.shape
```

```python
#Agrupación del nuevo dataframe (df_higher_90) en preparacion del test
df_higher_90_prepared = df_higher_90.groupby('test preparation course').size()
#Reset index y renombrar columna de la cantidad
df_higher_90_prepared = df_higher_90_prepared.reset_index().rename(columns={0: 'Cantidad'})
```

```python
#Porcentaje de estudianes con nota superior a 90 con base al total de estudiantes. 
(df_higher_90_prepared['Cantidad'].sum() / len(df))*100
```

```python
#Porcentaje que tuvieron una nota superior a 90 en las tres evaluaciones y que se prepararon, con base al total de estudiantes. 
df_higher_90_prepared['Porc_del_total'] = ( df_higher_90_prepared['Cantidad'] / df_higher_90_prepared['Cantidad'].sum())*100
df_higher_90_prepared
```
